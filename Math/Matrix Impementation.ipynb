```python
identity_matrix = np.eye(3, 4, dtype = int)
print(identity_matrix)
```

Transpose of a matrix

```python
matrix = np.random.randint(100, size = (4,5), dtype = int)
print(matrix)
```

```python
print(matrix[ : , 0])
```

```python
transpose_matrix = np.transpose(matrix)
print(transpose_matrix)
```

```python
print(transpose_matrix[ : , 0])
```

matrix addition

```python
a = np.array([ [2,3] , [4,5] ])
b = np.array([ [6,7] , [8,9] ])

a_plus_b = a++b

print(a_plus_b)
```

```python
sum = np.add(a,b)
print(sum)
```

```python
sum = np.subtract(a,b)
print(sum)
```
