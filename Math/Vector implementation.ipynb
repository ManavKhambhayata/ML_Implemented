**Plotting a vector**

```python
plt.quiver(0,0,4,2)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units = 'xy', angles = 'xy',  scale = 1, color = 'b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

**addition of two vectors**

```python
vector_1 = np.asarray([0,0,2,3])

vector_2 = np.asarray([0,0,3,-2])

sum = vector_1 + vector_2
print(sum)
```

**Dot Product**

```python
a = np.array([2,3])

b = np.array([4,4])

dotproduct = np.dot(a,b)

print(dotproduct)
```

```python
c = np.array([2,3,9])

d = np.array([4,4,7])

dotproduct2 = np.dot(c,d)

print(dotproduct2)
```

**CROSS PRODUCT**

```python
a = np.array([2,3,5])

b = np.array([4,4,2])

crossproduct = np.cross(a,b)

print(crossproduct)
```

**projectiojn of vector**

```python
a = np.array([2, 5])   

v = np.array([8, -6])  
  
# Task: Project vector u on vector v
  
# finding magn of the vector v

magnitude_of_v = np.sqrt(sum (v**2) )    
  

proj_of_a_on_v = (np.dot(a, v)/magnitude_of_v**2)*v
  
print("Projection of Vector u on Vector v is: ", proj_of_a_on_v)
```
